### Approaches to Handling Missing `LotFrontage` Data

Since zero was **not originally included** in the data, there are two main ways I Could approach handling the missing values:

#### Treat Missingness as MNAR (Missing Not at Random)
- Interpretation: The missing `LotFrontage` values may indicate that **some houses genuinely have no measurable frontage** (e.g., corner or irregular lots).
- In this case, the missingness is **related to the unobserved value itself**  houses with no frontage tend to have it missing.
- **Possible approach:**  
  - Fill missing values with an **arbitrary value (e.g., 0)** to explicitly represent *no frontage*.  
  - However, I could only do this if the guys upstairs told me that *missing = no frontage*.

#### Treat Missingness as MAR (Missing At Random)
- Interpretation: The missingness in `LotFrontage` is likely **related to other observed features**, such as:
  - `Neighborhood`
  - `HouseStyle`
  - `LotArea`
- In this case, the missing values depend on *observed data*, not on the missing value itself.
- **Possible approach:**  
  - Impute missing values using neighborhood-level, approaching using the median

```python
# To see if certain neighborhoods tend to have more missing LotFrontage values.
housing_missing_data.groupby('Neighborhood')['LotFrontage'].apply(lambda x: x.isnull().mean())
```

```python
# To see if certain property types (e.g., condos, townhomes, 1-story houses) dont have frontage.
housing_missing_data.groupby('HouseStyle')['LotFrontage'].apply(lambda x: x.isnull().mean())
```

```python
# To check if missing frontage correlates with smaller or irregular lot sizes
housing_missing_data.groupby(housing_missing_data['LotFrontage'].isnull())['LotArea'].describe()
```

```python
sns.boxplot(x=housing_missing_data['LotFrontage'].isnull(), y=housing_missing_data['LotArea'])
```

```python
# this imputes the missing values using the median of each neighborhood.
housing_missing_data['LotFrontage'] = (
      housing_missing_data.groupby('Neighborhood')['LotFrontage']
      .transform(lambda x: x.fillna(x.median()))
  )
```

```python
# shows that there are no longer missing values
housing_missing_data['LotFrontage'].isnull().sum()
```

### TASK 4 - Statistical Imputation for Outliers

#### I'm going to subset a few columns relevant to sale price

```python
housing_missing_data.info()
```

```python
# subsets the data
cols = ['SalePrice', 'LotArea', 'GrLivArea', 'GarageArea', 'TotalBsmtSF']
house_subset = housing_missing_data[cols].copy()
```

```python
house_subset.head(10)
```

```python
Q1 = house_subset.quantile(0.25)
Q3 = house_subset.quantile(0.75)
IQR = Q3 - Q1

outliers = (house_subset < (Q1 - 1.5 * IQR)) | (house_subset > (Q3 + 1.5 * IQR))
outlier_counts = outliers.sum().sort_values(ascending=False)
outlier_counts
```

```python
for col in ['SalePrice', 'LotArea', 'GrLivArea', 'GarageArea', 'TotalBsmtSF']:
    plt.figure(figsize=(6, 4))
    sns.boxplot(x=house_subset[col])
    plt.title(f'Boxplot of {col}')
```

```python
# Remove extreme outliers for SalePrice, LotArea, and GrLivArea
cols_remove = ['SalePrice', 'LotArea', 'GrLivArea']

# Start with a clean copy
df_cleaned = house_subset.copy()

for col in cols_remove:
    Q1 = df_cleaned[col].quantile(0.25)
    Q3 = df_cleaned[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    # Keep only values within the IQR range
    df_cleaned = df_cleaned[(df_cleaned[col] >= lower) & (df_cleaned[col] <= upper)]

print(f"Data shape after removing outliers: {df_cleaned.shape}")

# Impute outliers for GarageArea and TotalBsmtSF
cols_impute = ['GarageArea', 'TotalBsmtSF']

for col in cols_impute:
    Q1 = df_cleaned[col].quantile(0.25)
    Q3 = df_cleaned[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    median_val = df_cleaned[col].median()

    # Replace outliers with median
    df_cleaned[col] = df_cleaned[col].apply(
        lambda x: median_val if (x < lower or x > upper) else x
    )
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

cols = ['SalePrice', 'LotArea', 'GrLivArea', 'GarageArea', 'TotalBsmtSF']

fig, axes = plt.subplots(nrows=len(cols), ncols=1, figsize=(8, 4 * len(cols)))

for i, col in enumerate(cols):
    sns.boxplot(x=df_cleaned[col], ax=axes[i], color="skyblue")
    axes[i].set_title(f'After Cleaning: {col}')
    axes[i].grid(True, axis='x', linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
```

### TASK 7 - Introductory Inferential Statistics

### Hypotheses

- **Null Hypothesis (H0):** The average house price  \$180,000  
- **Alternative Hypothesis (H1):** The average house price > \$180,000


```python
from scipy import stats

# Extract SalePrice as a numpy array
saleprice = house_subset['SalePrice'].values
```

```python
# Perform one-sample t-test against 180,000
t_stat, p_value = stats.ttest_1samp(saleprice, 180000)

print(f"T-statistic: {t_stat:.3f}")
print(f"P-value: {p_value/2:.5f}")  # divide by 2 for one-tailed test
```

### Interpretation

- The t-test provides statistical evidence to compare our sample mean with a hypothesized population mean.
- This approach moves beyond description (EDA) to **inference**, helping us make data-driven claims.

#### Decision Rule

- Significance level: ** = 0.05**  
- If p-value < 0.05  reject null hypothesis (H0)  
- If p-value  0.05  fail to reject H0

- The **t-statistic** measures how far the sample mean is from \$180,000 in units of standard error.  
- The **p-value** tells us the probability of observing our data if the true mean were \$180,000.

#### Results

- **T-statistic:** 0.443  
- **One-tailed p-value:** 0.3289  

#### Conclusion: Since p-value = 0.32889 > 0.05 :
- We **fail to reject the null hypothesis, H0**, this means there isnt enough evidence to claim the mean sale price is gretaer than \$180,000.
- The t-statistic is small (0.443), meaning the sample mean is **very close to 180,000**, and the observed difference is **not statistically significant**, this indicates that sale prices are roughly centered around this value in our dataset.
